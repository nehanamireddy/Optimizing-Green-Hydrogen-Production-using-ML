```python
import matplotlib.pyplot as plt

# Define the heights for which we have data
heights = [20, 40, 100, 200, 300, 500, 1000]

# Compute the mean for each variable at each height using list comprehensions
wind_speed_means = [data[f"wind speed at {h}m (m/s)"].mean() for h in heights]
wind_direction_means = [data[f"wind direction at {h}m (deg)"].mean() for h in heights]
tke_means = [data[f"turbulent kinetic energy at {h}m (m2/s2)"].mean() for h in heights]

# Create a figure with three subplots for comparison
plt.figure(figsize=(18, 5))

# Plot Mean Wind Speed vs. Height
plt.subplot(1, 3, 1)
plt.plot(heights, wind_speed_means, marker='o', linestyle='-', color='blue')
plt.xlabel("Height (m)")
plt.ylabel("Mean Wind Speed (m/s)")
plt.title("Mean Wind Speed vs Height")
plt.grid(True)

# Plot Mean Wind Direction vs. Height
plt.subplot(1, 3, 2)
plt.plot(heights, wind_direction_means, marker='o', linestyle='-', color='orange')
plt.xlabel("Height (m)")
plt.ylabel("Mean Wind Direction (deg)")
plt.title("Mean Wind Direction vs Height")
plt.grid(True)

# Plot Mean TKE vs. Height
plt.subplot(1, 3, 3)
plt.plot(heights, tke_means, marker='o', linestyle='-', color='green')
plt.xlabel("Height (m)")
plt.ylabel("Mean TKE (m²/s²)")
plt.title("Mean TKE vs Height")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reshape data for plotting:
# 1) Create a new DataFrame with columns: 'Height', 'WindSpeed'
plot_data = []
heights = [20, 40, 100, 200, 300, 500, 1000]

for h in heights:
    col_name = f"wind speed at {h}m (m/s)"
    # Create a temporary DataFrame with a 'Height' column and 'WindSpeed' column
    temp_df = pd.DataFrame({
        'Height': [h]*len(data),
        'WindSpeed': data[col_name]
    })
    plot_data.append(temp_df)

plot_data = pd.concat(plot_data, ignore_index=True)

# Now make a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Height', y='WindSpeed', data=plot_data)
plt.title('Wind Speed Distribution by Height')
plt.xlabel('Height (m)')
plt.ylabel('Wind Speed (m/s)')
plt.show()
```

```python
# Define a function to categorize time into parts of the day
def categorize_part_of_day(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    else:
        return "Night"

# Apply the function to create a new column
data["part_of_day"] = data.index.hour.map(categorize_part_of_day)

# Verify the new column
print(data[["wind speed at 100m (m/s)", "part_of_day"]].head())
```

```python
import pandas as pd
import numpy as np

# Constants for wind power calculation
air_density = 1.225  # kg/m³
turbine_radius = 40  # meters (adjust for your turbine)
rotor_area = np.pi * (turbine_radius**2)  # Area in m²

# Load dataset (Assuming data is already in a DataFrame named 'data')
# If timestamp is not already in datetime format, convert it
data.index = pd.to_datetime(data.index)

# Convert wind speed at 100m height to power output
data["wind_power"] = 0.5 * air_density * rotor_area * (data["wind speed at 100m (m/s)"] ** 3)

# Normalize wind direction using sine and cosine transformation
data["wind_dir_sin"] = np.sin(np.radians(data["wind direction at 100m (deg)"]))
data["wind_dir_cos"] = np.cos(np.radians(data["wind direction at 100m (deg)"]))

# Extract time-based features
data["hour"] = data.index.hour
data["month"] = data.index.month
data["day_of_week"] = data.index.dayofweek

# Create lag features for wind speed (past 1, 3, 6 hours)
data["wind_speed_lag_1h"] = data["wind speed at 100m (m/s)"].shift(1)
data["wind_speed_lag_3h"] = data["wind speed at 100m (m/s)"].shift(3)
data["wind_speed_lag_6h"] = data["wind speed at 100m (m/s)"].shift(6)

# Drop NaN values from shifting
data.dropna(inplace=True)

# Check the processed data
print(data.head())
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target variable (y)
features = ["wind speed at 100m (m/s)", "wind_dir_sin", "wind_dir_cos",
            "turbulent kinetic energy at 100m (m2/s2)", "hour", "month",
            "wind_speed_lag_1h", "wind_speed_lag_3h", "wind_speed_lag_6h"]
target = "wind_power"

# Split the data
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, shuffle=False)

# Check dataset sizes
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Train a Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on test data
y_pred = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Mean Absolute Error: {mae:.2f} kW")
print(f"Root Mean Squared Error: {rmse:.2f} kW")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label="Actual Wind Power", color="blue")
plt.plot(y_test.index, y_pred, label="Predicted Wind Power", color="red", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Wind Power (kW)")
plt.title("Actual vs Predicted Wind Power")
plt.legend()
plt.grid(True)
plt.show()
```
